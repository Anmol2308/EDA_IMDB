```python
def convert_list(l1):
  return eval(l1) 
```

```python
df.loc[~df['cast'].isna(), 'cast'] = df.loc[~df['cast'].isna(), 'cast'].apply(convert_list)  #removing all the null values
```

```python
df.loc[df['budget']<=1000, 'budget'] = df['budget'].median()                     #budget of no movie will be less than 1000 dollars so taking the median in that case
df.loc[df['revenue']<=1000, 'revenue'] = df['revenue'].median()                  #revenue of no movie will be less than 1000 dollars so taking the median in that case
df['profit'] = df['revenue'] - df['budget']
```

```python
df['roi'] = df['profit']/df['budget']*100
```

```python
def get_list_of_producers(list_of_crew):
  producer_list = []
  for crew in list_of_crew:
    if crew['job'] == 'Producer':
      producer_list.append(crew['name'])
  return producer_list
```

```python
def get_director(list_of_crew):
  producer_list = []
  for crew in list_of_crew:
    if crew['job'] == 'Director':
      return crew['name']
```

```python
df_list_dir_prod = df[~df['crew'].isna()]
```

```python
df_list_dir_prod['list_of_producer'] = df_list_dir_prod['crew'].apply(get_list_of_producers)
df_list_dir_prod['Director'] = df_list_dir_prod['crew'].apply(get_director)
```

```python
df_list_dir_prod[['title','list_of_producer', 'Director', 'roi']].sort_values('roi', ascending = False).iloc[0:3,:]
```
