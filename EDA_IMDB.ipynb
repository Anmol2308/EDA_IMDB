```python
def get_list_of_producers(list_of_crew):
  producer_list = []
  for crew in list_of_crew:
    if crew['job'] == 'Producer':
      producer_list.append(crew['name'])
  return producer_list
```

```python
def get_director(list_of_crew):
  producer_list = []
  for crew in list_of_crew:
    if crew['job'] == 'Director':
      return crew['name']
```

```python
df_list_dir_prod = df[~df['crew'].isna()]
```

```python
df_list_dir_prod['list_of_producer'] = df_list_dir_prod['crew'].apply(get_list_of_producers)
df_list_dir_prod['Director'] = df_list_dir_prod['crew'].apply(get_director)
```

```python
df_list_dir_prod[['title','list_of_producer', 'Director', 'roi']].sort_values('roi', ascending = False).iloc[0:3,:]
```

```python
df_cast = df[~df['cast'].isnull()]
```

```python
freq_dict = {}
def create_freq_dict(row):
  for cast in row:
    name = cast['name']
    if name in freq_dict:
      freq_dict[name] = freq_dict[name]+1
    else:
      freq_dict[name] = 1
```

```python
df_cast['cast'].apply(create_freq_dict)
```

```python
actor_sorted_dict = dict(sorted(freq_dict.items(), key=lambda item: item[1], reverse = True))
```
